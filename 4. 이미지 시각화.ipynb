## package 불러오기

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

---

## 이미지 파일 열기

opencv로도 열수 있지만, shape 순서가 바뀔 때도 있어서 PIL로 여는 것을 개인적으로 선호

```python
path = 'source/dog.jpg'  # 절대경로로 'C:/Workspace/Fast Campus/source/dog.jpg'로 줄 수 있음

image_pil = Image.open(path)
image = np.array(image_pil)
```

---

## 이미지 들여다 보기 

```python
image.shape
```

```python
np.max(image), np.min(image)
```

이미지를 열기 전에 shape 및 min, max를 통해서 이미지의 range 확인 필요

## 그래프로 시각화 하기

```python
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

---

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

---
## 이미지 흑백으로 열기

```python
image_bw = image_pil.convert('L')
image_bw = np.array(image_bw)
image_bw.shape  
```

---
## 흑백 이미지 열기

```python
# 2개의 차원일 때는 gray scale로 열림
plt.imshow(image_bw)
plt.show()
```

---
